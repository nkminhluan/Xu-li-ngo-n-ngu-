# Lấy dữ liệu tạo DataFrame để train model CRF

```python
import re
tag_train = re.findall(r'(\w+|\s+|\...|[A-Za-z-“”:()?;%>><</~*&^.,!-]+)'+r'([/]\w+|[/]\-|[/]\,|[/]\.)',str(f_train1))
```

```python
tag_train
```

```python
import pandas as pd
data = pd.DataFrame(tag_train, columns=['Word', 'Tag']) 
data
```

# Tạo mảng chứa các từ trong câu

```python
import numpy as np
sentences=[]
i=0
while i<len(data):
    tmp=[]
    while (data.iloc[i,0])!="." and data.iloc[i,1] !='.':
        tmp.append(data.iloc[i,0])
        i=i+1
    tmp.append('.')
    sentences.append(np.array(tmp))
    i=i+1
```

```python
sentences
```

# Tạo mảng lấy các nhãn của các từ

```python
tagged_sentences=[]
i=0
while i<len(data):
    tmp=[]
    while (data.iloc[i,0])!="." and data.iloc[i,1] !='.':
        tmp.append(data.iloc[i,1])
        i=i+1
    tmp.append('.')
    tagged_sentences.append(np.array(tmp))
    i=i+1
```

```python
tagged_sentences
```

```python
from sklearn.model_selection import train_test_split
train_sentences, test_sentences, train_tags, test_tags = train_test_split(sentences, tagged_sentences, test_size=0.2)
```

# Train với model có Features

```python
def features(sentence,index):
    return {
        'is_first_capital':int(sentence[index][0].isupper()),
        'is_first_word': int(index==0),
        'is_last_word':int(index==len(sentence)-1),
        'is_complete_capital': int(sentence[index].upper()==sentence[index]),
        'prev_word':'' if index==0 else sentence[index-1],
        'next_word':'' if index==len(sentence)-1 else sentence[index+1],
        'is_numeric':int(sentence[index].isdigit()),
        'is_alphanumeric': int(bool((re.match('^(-?=.*[0-9]$)(?=.*[a-zA-Z])',sentence[index])))),
        'prefix_1':sentence[index][0],
        'prefix_2':sentence[index][:2],
        'prefix_3':sentence[index][:3],
        'prefix_4':sentence[index][:4],
        'suffix_1':sentence[index][-1],
        'suffix_2':sentence[index][-2:],
        'suffix_3':sentence[index][-3:],
        'suffix_4':sentence[index][-4:],
        'word_has_hyphen': 1 if '-' in sentence[index] else 0,
        
        
    }
```

```python
def prepareData(tagged_sentences):
    X=[]
    for sentences in tagged_sentences:
        X.append([features(sentences, index) for index in range(len(sentences))])
    return X
```

```python
X_traincrf = prepareData(train_sentences)
y_traincrf = train_tags
X_testcrf = prepareData(test_sentences)
y_testcrf =  test_tags
```

```python
crf1 = CRF(algorithm='lbfgs')
model1  = crf1.fit(X_traincrf, y_traincrf)
```

```python
insen1 = 'Trên đường xuất_hiện nhiều cặp cha - con , mẹ - con hay có khi là cả nhà tay_xách_nách_mang .'
crfpre1 = nltk.word_tokenize(insen1)
print(crfpre1)
print(model1.predict([crfpre1])[0])
final_list1 = [crfpre1[i] + model1.predict([crfpre1])[0][i] for i in range(len(crfpre1))]  
print(final_list1)
final1 = ' '.join([str(elem) for elem in final_list1])
print(final1)
```

```python
from sklearn_crfsuite import metrics
 
y_predfe = model1.predict(X_testcrf)
print('Accuray with features: ',metrics.flat_accuracy_score(y_testcrf, y_predfe))
 
```
